```python
import cx_Oracle
import pandas as pd
import win32com.client, gc
from time import sleep

## Declaration of variables
ip = 'your_ip_address'
port = 'your_port_number'
service_name = 'your_service_name'
username = 'your_username'
password = 'your_password'
sql_script = 'your_sql_query_here'

path_data_raw = 'output_file.xlsx'  
path_report = 'report_file.xlsx'

# Main script
dsn_tns = cx_Oracle.makedsn(ip, port, service_name=service_name)
conn = cx_Oracle.connect(user=username, password=password, dsn=dsn_tns).cursor()

query_output = conn.execute(sql_script)

title = [i[0] for i in conn.description]
df = pd.DataFrame(data = query_output, columns = title)
df.to_excel(path_data_raw, index=False)

xl=win32com.client.DispatchEx("Excel.Application")
xl.Application.Visible = False      ## Open in background. should be True for debugging
xl.Application.DisplayAlerts = False    ## to avoid any pop ups
wb = xl.Workbooks.Open(path_report)
sleep(3)     ### to wait application load. 3 is number of seconds to wait. increase if needed
wb.RefreshAll()    ## refresh all in excel
xl.CalculateUntilAsyncQueriesDone()  
sleep(2)                            ## to ensure all things is done
print('Done Refresh')
wb.Save()
sleep(2)
xl.Application.Quit()
xl = None
del xl
gc.collect()
```
